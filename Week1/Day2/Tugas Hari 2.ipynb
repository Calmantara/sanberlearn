<h2>Soal 2: Built-in Function</h2>

Jawab Pertanyaan di bawah ini:
    
Apa itu built-in Function?

Sebutkan 3 built-in Function di pyhon!

Berikan contoh masing2 1 cara penggunaan built-in function yang kamu sebutkan sebelumnya

//isi jawaban text disini

Function that has already defined in python as Object. So we can call it whenever we need without any external module/library 

```python
# Contoh penggunaan built in function
arr = [1, 2, 3, 4]
# len
print("(len) array length: ", len(arr))
# max
print("(max) max element of array: ", max(arr))
# map
print("(map) array element plus 2: ", list(map(lambda x: x + 2, arr)))
```

<h2>Soal 3: Method dan Function</h2>

Jawab Pertanyaan di bawah ini:
    
- Apa perbedaan method dan function?

Method is function that object' have. For example, String object has its own method such as upper, split, join, etc. List object has its own method such as append, clear, remove, etc.

A function is a block of code which only runs when it is called. We can pass data, known as parameters, into a function. A function can return data as a result.

<h2>Soal 4: Menggunkan Method String</h2>

Lengkapi kode untuk menghasilkan suatu output yang di harapkan

```python
kalimat = "Corona cepat selesai"

# gunakan method untuk mengubah nilai kalimat menjadi uppercase semua kemudian tampilkan hasilnya
print("Uppercase: {}".format(kalimat.upper()))

# gunakan method untuk menghitung berapa huruf e di dalam kalimat
print("Amount of e char: {}".format(kalimat.count("e")))
```

Expected Output:

CORONA CEPAT SELESAI

3

<h2>Soal 5: Menggunkan Method List</h2>

Lengkapi kode untuk menghasilkan suatu output yang di harapkan

```python
areas = [11.25, 18.0, 20.0, 10.75, 9.50]

# gunakan suatu method dari object list untuk menampilkan index dari nilai 20
print("Index of element 20 is {}".format(areas.index(20)))

# gunakan suatu method dari object list untuk menambahkan nilai 15.5 kedalam list ke index terakhirnya
areas.append(15.5)
print(areas)
```

<h2>Soal 6: Membuat Simple Function</h2>

Buatlah suatu fungsi yang menerima satu input argumen berbentuk list dan mempunyai elemen bertipe numeric semua, dimana fungsi tersebut berguna untuk menghitung rata2 dari kumpulan elemen list tersebut. namai fungsi tersebut 'mean_list'

```python
# Soal 6
print("\nSoal 6")
obj_list = [11.25, 18.0, 20.0, 10.75, 9.50]


def mean_list(inp_list: list) -> float:
    return sum(inp_list) / len(inp_list)


print(mean_list(obj_list))
```

Expected Output:

13.9

<h2>Soal 7: Membuat Function dengan Multiple arguments</h2>

Buatlah suatu fungsi untuk melakukan penggabungan antara dua list

```python
print("\nSoal 7")
obj_list = [2, 4, 5, 6]
obj_pengali = [1, 2, 3]


def kali_list(in_list: list, times_list: list) -> list:
    return in_list + times_list


print(kali_list(obj_list, obj_pengali))
```

Expected Output:

[2, 4, 5, 6, 1, 2, 3]
