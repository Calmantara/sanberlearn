<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Jawab: 

1. Series is one dimention object that can store many types of data. The difference with array is, in series we can add custom index such as 'a', 'b', 'c', to access the data. But, by default the index is numerical increment from zero, so in dictionary analogy, index is key. The different is dictionary can not be created if doesn't have key, series can be created even we don't set the index

2. DataFrame is 2 dimentions object, like table. Series just 1 dimention object  

```python
# code here
x = pd.Series([1,'a',2], index=['a', 'b', 'c'])

print('Series:\n', x)
print('Access Series with number index:', x[1])
print('Access Series with number index:', x['a'])
```

<h2>Soal 6: Practice Deskriptif Statistik</h2>

- Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
- Berapa nilai rata-rata dari column price
- Berapa nilai standar deviasi dari column depth
- Berapa nilai maximum dari column carat

```python
# code here
ds = df.describe()
print('Descriptive Statistic')
print(ds)

print('Mean of price column', ds['price']['mean'])
print('Std of depth column', ds['depth']['std'])
print('Max of carat column', ds['carat']['max'])
```
