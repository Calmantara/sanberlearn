```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
max_tan = max(tan)
max_sin = max(sin)
max_cos = max(cos)
max_lim = max_tan if max_tan > max_sin else max_sin
max_lim = max_lim if max_lim > max_cos else max_cos

min_tan = min(tan)
min_sin = min(sin)
min_cos = min(cos)
min_lim = min_tan if min_tan < min_sin else min_sin
min_lim = min_lim if min_lim < min_cos else min_cos

fig, axis = plt.subplots(ncols=3)

fig.set_figheight(8)
fig.set_figwidth(20)

axis[0].plot(x, tan, color='blue', label='tanX')
axis[0].set_xlabel('X-Axis Label')
axis[0].set_ylabel('Y-Axis Label')
axis[0].set_title('Grafik tanx')
axis[0].set_ylim(min_lim, max_lim)

axis[1].plot(x, sin, color='red', label='sinX')
axis[1].set_xlabel('X-Axis Label')
axis[1].set_ylabel('Y-Axis Label')
axis[1].set_title('Grafik sinx')
axis[1].set_ylim(min_lim, max_lim)

axis[2].plot(x, cos, color='green', label='cosX')
axis[2].set_xlabel('X-Axis Label')
axis[2].set_ylabel('Y-Axis Label')
axis[2].set_title('Grafik cosx')
axis[2].set_ylim(min_lim, max_lim)

fig.legend(loc='upper center')
```
