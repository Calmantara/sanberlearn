Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
import matplotlib.pyplot as plt

fig, axis = plt.subplots()

fig.set_figheight(7)
fig.set_figwidth(12)


axis.plot(x, cos, color='red', label='Cosx')
axis.plot(x, cos3, color='green', linestyle='--', label='3 Cosx')
axis.plot(x, sin, color='blue', linestyle='-.', label='Sin(x-90)')
axis.plot(x, sin2, color='yellow', linestyle='', marker='x', label='2 Sinx')

axis.set_xlabel('X-Axis Label')
axis.set_xlabel('Y-Axis Label')
axis.set_title('Grafik Cos Sin')

axis.legend(loc="center")
```
